# load libraies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
```

# load_dataset

```python
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()
cancer_data
```

```python
cancer_df = pd.DataFrame(np.c_[cancer_data["data"],cancer_data["target"]],
                         columns = np.append(cancer_data["feature_names"], ["target"]))
cancer_df.shape
```

# split data

```python
X = cancer_df.iloc[:, 0:-1]
X.shape
```

```python
y = cancer_df.iloc[:, -1]
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=123)
```

```python
print("shape of X_train", X_train.shape)
print("shape of y_train", y_train.shape)
print("shape of X_test", X_test.shape)
print("shape of y_test", y_test.shape)
```

# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train, y_train)
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

# Model building

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# Support vactor classifier

```python
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)
y_pred = svc_classifier.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
# scalling data
svc_classifier2 = SVC()
svc_classifier2.fit(X_train_sc, y_train)
y_pred_sc = svc_classifier2.predict(X_test_sc)
accuracy_score(y_pred_sc, y_test)
```

# decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc_classifier = DecisionTreeClassifier(criterion='entropy', random_state=123)
dtc_classifier.fit(X_train, y_train)
y_pred_dtc = dtc_classifier.predict(X_test)
accuracy_score(y_pred_dtc, y_test)
```

```python
# scalling data
dtc_classifier2 = DecisionTreeClassifier(criterion='entropy', random_state=123)
dtc_classifier2.fit(X_train_sc, y_train)
y_pred_dtc_sc = dtc_classifier2.predict(X_test_sc)
accuracy_score(y_pred_dtc_sc, y_test)
```

# Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc_classifier = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=123)
rfc_classifier.fit(X_train, y_train)
y_pred_rfc = rfc_classifier.predict(X_test)
accuracy_score(y_pred_rfc, y_test)
```

```python
y_pred_rfc
```

```python
# scaller data
rfc_classifier2 = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=123)
rfc_classifier2.fit(X_train_sc, y_train)
y_pred_rfc_sc = rfc_classifier2.predict(X_test_sc)
accuracy_score(y_pred_rfc_sc, y_test)
```

# KNN classifier(K nearest neighbor)

```python
from sklearn.neighbors import KNeighborsClassifier
knc_classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knc_classifier.fit(X_train, y_train)
y_pred_knc = knc_classifier.predict(X_test)
accuracy_score(y_pred_knc, y_test)
```

```python
# scaller data
knc_classifier2 = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knc_classifier2.fit(X_train_sc, y_train)
y_pred_knc_sc = knc_classifier2.predict(X_test_sc)
accuracy_score(y_pred_knc_sc, y_test)
```

# naive bayes classifier

```python
from sklearn.naive_bayes import GaussianNB
gnb_classifier = GaussianNB()
gnb_classifier.fit(X_train, y_train)
y_pred_gnb = gnb_classifier.predict(X_test)
accuracy_score(y_pred_gnb, y_test)
```

```python
# scalling data
gnb_classifier2 = GaussianNB()
gnb_classifier2.fit(X_train_sc, y_train)
y_pred_gnb_sc = gnb_classifier2.predict(X_test_sc)
accuracy_score(y_pred_gnb_sc, y_test)
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression(penalty='l2',C=1.0, random_state=123)
lr_classifier.fit(X_train,y_train)
y_pred_lr = lr_classifier.predict(X_test)
accuracy_score(y_pred_lr, y_test)
```

```python
# scaller data
lr_classifier2 = LogisticRegression(penalty='l2',C=1.0, random_state=123)
lr_classifier2.fit(X_train_sc,y_train)
y_pred_lr_sc = lr_classifier2.predict(X_test_sc)
accuracy_score(y_pred_lr_sc, y_test)
```

# adaboostclassifier

```python
from sklearn.ensemble import AdaBoostClassifier
ada_classifier = AdaBoostClassifier(n_estimators=50, learning_rate=1.0, algorithm='SAMME.R', random_state=123)
ada_classifier.fit(X_train, y_train)
y_pred_ada = ada_classifier.predict(X_test)
accuracy_score(y_pred_ada, y_test)
```

# Cross-validation of the model

```python
from sklearn.model_selection import cross_val_score
cross_validation = cross_val_score(estimator=rfc_classifier, X=X_train, y=y_train, cv=10)
print("cross validation accuracy of random forest model = ", cross_validation)
print("\ncross validation mean accuracy of random forest model = ", cross_validation.mean())
```

```python
import joblib

# save the model
joblib.dump(rfc_classifier, "breast_cancer_detector_joblib")

# load the model
breast_cancer_detector_joblib_model = joblib.load("breast_cancer_detector_joblib")

# predict data
y_pred = breast_cancer_detector_joblib_model.predict(X_test)

# check confusion metrix
print("confusion metrix= ", confusion_matrix(y_test, y_pred))

# check model accuracy
print("accuracy of model= ", accuracy_score(y_test, y_pred))
```
